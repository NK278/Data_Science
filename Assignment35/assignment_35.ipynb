- Dataset Source - https://www.kaggle.com/datasets/uciml/red-wine-quality-cortez-et-al-2009/data
- The data consists of 12 column and 1599 rows.
- The "wine quality" dataset typically includes various features that describe different aspects of wine, and the quality of the wine is often rated by experts or through sensory evaluations. While the specific features can vary, common features in such datasets often include:

1. **Fixed Acidity:** The amount of non-volatile acids in the wine. These acids contribute to the overall acidity of the wine, affecting its taste and stability.

2. **Volatile Acidity:** The amount of volatile acids in the wine. Too much volatile acidity can lead to an unpleasant vinegar-like taste.

3. **Citric Acid:** Citric acid can add freshness and flavor to the wine. It is a component that contributes to the overall acidity.

4. **Residual Sugar:** The amount of sugar remaining after fermentation stops. Wines with higher residual sugar can taste sweeter.

5. **Chlorides:** The amount of salt in the wine. It is usually measured as sodium chloride and can impact the wine's taste and mouthfeel.

6. **Free Sulfur Dioxide:** Sulfur dioxide is used as a preservative in winemaking. It prevents the wine from spoiling and acts as an antioxidant.

7. **Total Sulfur Dioxide:** The total amount of sulfur dioxide in the wine, including both free and bound forms. High levels can affect the taste and smell of the wine.

8. **Density:** The density of the wine. It is influenced by factors such as sugar content and alcohol concentration.

9. **pH:** The pH level indicates the acidity or basicity of the wine. It can impact the taste, color, and stability of the wine.

10. **Sulphates:** The amount of sulfates in the wine. Sulfates can contribute to the wine's antioxidant properties.

11. **Alcohol:** The alcohol content of the wine. It can affect the body, mouthfeel, and overall balance of the wine.

These features are essential in predicting the quality of wine because they influence various sensory attributes, such as taste, aroma, and mouthfeel. The balance and interaction of these features contribute to the overall character of the wine. For example, a well-balanced acidity, appropriate sugar levels, and the right alcohol content can result in a harmonious and high-quality wine. Additionally, the presence of preservatives like sulfur dioxide can affect the wine's aging potential and stability. Analyzing these features collectively allows for the development of models that can predict the quality of wine based on its chemical composition.

```python
# Visualize the distribution of each feature using histograms
df.hist(bins=20, figsize=(15, 12))
plt.suptitle('Distribution of Features in Wine Quality Dataset')
plt.show()

# Use seaborn to create kernel density plots for each feature
plt.figure(figsize=(15, 12))
for i, column in enumerate(df.columns):
    plt.subplot(3, 4, i + 1)
    sns.kdeplot(df[column], fill=True)
    plt.title(f'Distribution of {column}')

plt.tight_layout()
plt.show()

# Check for normality using the Shapiro-Wilk test
for column in df.columns:
    stat, p_value = shapiro(df[column])
    print(f"Shapiro-Wilk test for {column}: p-value = {p_value}")

# Check skewness
skewness = df.apply(lambda x: skew(x))
print("\nSkewness of each feature:")
print(skewness)
```

# Apply Box-Cox transformation to features with non-normal distributions
for column in df.columns:
    transformed_data, _ = boxcox(df[column] + 1)  # Adding 1 to handle zero values
    plt.figure(figsize=(10, 4))

    sns.kdeplot(df[column], fill=True)
    plt.title(f'Original Distribution of {column}')

    plt.subplot(1, 2, 2)
