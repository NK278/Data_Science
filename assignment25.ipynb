Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
import plotly.express as px

# Load the Titanic dataset
titanic_data = sns.load_dataset("titanic")

# Create a scatter plot using Plotly Express
scatter_plot = px.scatter(titanic_data, x="age", y="fare", title="Scatter Plot of Age vs Fare")
scatter_plot.show()
```

Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
import seaborn as sns
import plotly.express as px

# Load the "tips" dataset from Seaborn
tips_data = sns.load_dataset("tips")

# Create a box plot using Plotly Express
box_plot = px.box(tips_data, x="day", y="total_bill", title="Box Plot of Total Bill Amounts by Day")
box_plot.show()
```

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
import seaborn as sns
import plotly.express as px

# Load the "tips" dataset from Seaborn
tips_data = sns.load_dataset("tips")

# Create a histogram with customizations using Plotly
histogram = px.histogram(
    tips_data,
    x="total_bill",
    color="day",
    facet_col="smoker",
    facet_col_wrap=2,
    title="Histogram of Total Bill Amounts by Day and Smoker"
)

histogram.update_traces(opacity=0.7)  # Adjust opacity for better visualization
histogram.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset from seaborn
iris_data = sns.load_dataset("iris")

# Create a scatter matrix using seaborn's pairplot
sns.set(style="ticks")
pair_plot = sns.pairplot(
    iris_data,
    hue="species",
    markers=["o", "s", "D"],  # Custom markers for each species
    diag_kind="hist",        # Use histograms on diagonal axes
    plot_kws={"alpha": 0.7}   # Adjust point opacity for better visualization
)

plt.suptitle("Scatter Matrix Plot of Iris Dataset")
plt.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

Ans: distplot was a function provided by the seaborn library for creating distribution plots, but it has been deprecated in favor of using histplot for histograms and kdeplot for kernel density estimation plots. Plotly Express, on the other hand, doesn't have a direct equivalent for the deprecated distplot, but you can create similar plots using the histplot and kdeplot functions in seaborn or using the px.histogram function in Plotly Express.



```python
import seaborn as sns
import plotly.express as px

# Load example data
tips_data = sns.load_dataset("tips")

distplot = px.histogram(
    tips_data,
    x="total_bill",
    marginal="rug",  # Adds rug plot to the sides
    title="Distribution of Total Bill Amounts"
)

distplot.show()
```
