Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans:
1)istograms (sns.histplot()):
Use: Histograms are used to visualize the distribution of a single continuous variable. They display the frequency of data points falling into predefined bins or intervals.
Common Use Cases: Exploring data distribution, identifying outliers, assessing data skewness.

2)Scatter Plots (sns.scatterplot()):
Use: Scatter plots are used to visualize the relationship between two continuous variables. Each data point is represented as a point on the plot.
Common Use Cases: Identifying patterns, correlations, and clusters in data points, assessing the strength and direction of relationships.

3)Bar Plots (sns.barplot() or sns.countplot()):
Use: Bar plots display categorical data using rectangular bars. barplot() can show the mean or another summary statistic of a numerical variable for each category, while countplot() displays the count of occurrences of each category.
Common Use Cases: Comparing the values of a categorical variable, showing frequency counts, visualizing summary statistics across categories.

4)Box Plots (sns.boxplot()):
Use: Box plots (box-and-whisker plots) show the distribution of a numerical variable through its quartiles and any potential outliers. They provide insights into the central tendency and spread of data.
Common Use Cases: Identifying variability and skewness in data, visualizing the spread of data within different groups or categories.

5)Violin Plots (sns.violinplot()):
Use: Violin plots combine a box plot and a kernel density estimation plot to show the distribution of a continuous variable across different categories or groups.
Common Use Cases: Visualizing the distribution and spread of data within groups, comparing multiple distributions, identifying modes and skewness.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri_data = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
plt.title("FMRI Signal Over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
titanic_data = sns.load_dataset("titanic")
plt.figure(figsize=(10, 6))  
sns.boxplot(x="pclass", y="age", data=titanic_data)
plt.title("Age Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(10, 6))  
sns.boxplot(x="pclass", y="fare", data=titanic_data)
plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
diamonds_data = sns.load_dataset("diamonds")
plt.figure(figsize=(10, 6))  
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple='stack')
plt.title("Price Distribution by Cut of Diamond")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.legend(title="Cut")
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
iris_data = sns.load_dataset("iris")

# Create a pair plot with hue
sns.set(style="ticks")  # Set the style
sns.pairplot(iris_data, hue="species")
plt.suptitle("Pair Plot of Iris Dataset by Species", y=1.02)  # Add title above plot
plt.show()
```
